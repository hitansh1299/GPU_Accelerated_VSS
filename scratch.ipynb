```python
eight_bit_number = 0b00011111  # Replace this with your 8-bit number in binary format

# Extract the first 5 bits
first_five_bits = eight_bit_number & 0b11111
first_five_bits = eight_bit_number // 0b1000

print(f"The value of the first 5 bits is: {first_five_bits}")
```

```python
get_first_5_bits = lambda x: x // 0b1000
```

```python
pixel = (255,48,255)
gray = get_first_5_bits(pixel[0]) * 2048 + get_first_5_bits(pixel[1]) * 64 + get_first_5_bits(pixel[2]) * 2
print(bin(gray))
```
